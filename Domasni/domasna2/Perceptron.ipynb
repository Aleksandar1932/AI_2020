```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

fig = plt.figure()
fig.set_size_inches(10, 6)
ax = fig.add_subplot(111, projection='3d')

# ne: 0, da:1
# nok: 0, den: 1

# Opasnost: Niska
grad_n =   [0,0,0,1,1]
denNok_n = [0,1,1,1,1]
iten_n =   [1,0,0,1,1]

# Opasnost: Visoka
grad_v =   [0,0,1,1,1]
denNok_v = [0,0,0,0,1]
iten_v =   [0,1,0,0,1]


ax.scatter(grad_n, denNok_n, iten_n, c='g', marker='o', s=1000)
ax.scatter(grad_v, denNok_v, iten_v, c='r', marker='+', s=500)


ax.set_xlabel('')
ax.set_ylabel('/ ')
ax.set_zlabel(' ')


plt.show()
```
